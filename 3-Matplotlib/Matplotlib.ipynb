### Load Necessary Library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]        # y = 2x


# Resize the graph(dpi specifies pixcels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5),dpi=110)


# Line 1

# Keyword Argument Notation
# plt.plot(x,y,label='2x',color='red', linewidth = 3, linestyle='--', marker = '.', markersize=15,markeredgecolor='blue')

# use Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'b*-.',label='2x')


## Line 2

# select the intervals we want to plot points at
x2 = np.arange(0,4.5,0.5)
# print(x2)

# Plot part of the graph as Line 
plt.plot(x2[:6], x2[:6]**2, 'r',label='x^2')

# Plot part of the graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')


# Add a title(specify font parameters with fontdict)
plt.title('First Graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})


# X and Y lables
plt.xlabel('X Axis')
plt.ylabel('Y Axis')


# X, Y axis Tickmarks (Scale of our graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,1,4,6,8,10,12])


# Add a legend
plt.legend()


# Save figure(dpi 300 is good when saving so graph has high resolution)
plt.savefig("Matplotlib graph.png", dpi=300)


# Show plot
plt.show()                    
# to get rid off [<matplotlib.lines.Line2D at 0x26b8afe5a60>]

```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

barGraph = plt.bar(labels, values)


# adding different patterns to the each bars
barGraph[0].set_hatch('/')
barGraph[1].set_hatch('o')
barGraph[2].set_hatch('*')


plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

barGraph = plt.bar(labels, values)


patterns = ['/', 'o', '*']
for bar in barGraph:
    bar.set_hatch(patterns.pop(0))              
#     bar.set_hatch(patterns.pop())           # pop operation

plt.figure(figsize=(6,4))

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
PSG = fifa.loc[fifa.Club == 'Paris Saint-Germain']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC barcelona', 'madrid', 'PSG', 'NE Revolution']

boxes = plt.boxplot([barcelona, madrid, PSG, revs], labels = labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # setedge color
    box.set(color='#1e81b0', linewidth = 2)
    
    # fill color inside box
    box.set(facecolor = '#eeeee4')

plt.title('Professional Soccer Team Comparision')

plt.ylabel('FIFA Overall Rating')


plt.show()
```
